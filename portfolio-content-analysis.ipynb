```python
# Technical Skills Analysis & Visualization

# Create skills dataframe with proficiency levels
skills_data = []
skills = research_profile['technical_skills']

# Define skill categories and proficiency levels
skill_proficiency = {
    'Python': 95, 'PyTorch': 90, 'Computer Vision': 88, 'Remote Sensing': 85,
    'OpenCV': 85, 'NumPy': 90, 'Pandas': 88, 'scikit-learn': 85,
    'Medical Imaging': 80, 'Vision Transformers': 85, 'CNNs': 88,
    'Git': 85, 'SQL': 75, 'Jupyter': 90, 'Spatiotemporal Modeling': 85
}

for category, skill_list in skills.items():
    for skill in skill_list:
        skills_data.append({
            'skill': skill,
            'category': category.replace('_', ' ').title(),
            'proficiency': skill_proficiency.get(skill, 80),
            'years_experience': np.random.randint(2, 5)  # Estimated based on grad program
        })

# Add specializations
for skill in skills['specializations']:
    if skill not in [s['skill'] for s in skills_data]:
        skills_data.append({
            'skill': skill,
            'category': 'Specializations',
            'proficiency': skill_proficiency.get(skill, 80),
            'years_experience': np.random.randint(1, 3)
        })

skills_df = pd.DataFrame(skills_data)
print("ğŸ“Š Technical Skills Analysis")
print(skills_df.groupby('category')['proficiency'].agg(['mean', 'count']).round(1))
```

```python
# Create Skills Radar Chart for Portfolio
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{"type": "polar"}, {"type": "bar"}]],
    subplot_titles=("Technical Skills Proficiency", "Skills by Category")
)

# Top skills for radar chart
top_skills = skills_df.nlargest(8, 'proficiency')

# Radar chart
fig.add_trace(go.Scatterpolar(
    r=top_skills['proficiency'].tolist() + [top_skills['proficiency'].iloc[0]],
    theta=top_skills['skill'].tolist() + [top_skills['skill'].iloc[0]],
    fill='toself',
    fillcolor='rgba(0, 168, 225, 0.3)',  # Prime blue
    line=dict(color='rgb(0, 168, 225)', width=2),
    name="Proficiency"
), row=1, col=1)

# Category bar chart
category_stats = skills_df.groupby('category')['proficiency'].mean().sort_values(ascending=True)

fig.add_trace(go.Bar(
    x=category_stats.values,
    y=category_stats.index,
    orientation='h',
    marker_color='rgb(0, 168, 225)',
    name="Avg Proficiency"
), row=1, col=2)

# Update layout
fig.update_layout(
    height=400,
    title_text="Deepika Nuthalapati - Technical Skills Analysis",
    title_x=0.5,
    showlegend=False,
    paper_bgcolor='rgba(15, 23, 30, 1)',
    plot_bgcolor='rgba(15, 23, 30, 1)',
    font=dict(color='white', size=12)
)

# Update polar chart
fig.update_polars(
    radialaxis=dict(
        visible=True,
        range=[0, 100],
        gridcolor='rgba(255, 255, 255, 0.2)',
        gridwidth=1
    ),
    angularaxis=dict(
        gridcolor='rgba(255, 255, 255, 0.2)',
        gridwidth=1
    ),
    bgcolor='rgba(15, 23, 30, 0.8)'
)

# Update bar chart
fig.update_xaxes(range=[0, 100], row=1, col=2, gridcolor='rgba(255, 255, 255, 0.2)')
fig.update_yaxes(row=1, col=2, gridcolor='rgba(255, 255, 255, 0.2)')

fig.show()

print("ğŸ¯ Top Skills:")
for i, row in top_skills.iterrows():
    print(f"   {row['skill']}: {row['proficiency']}% proficiency")
```

```python
# Research Timeline & Publications Analysis

# Create timeline data
timeline_data = []

# Add education milestone
timeline_data.append({
    'date': '2024-12-01',
    'event': 'M.S. Computer Science Graduation',
    'type': 'Education',
    'importance': 10,
    'description': 'University of South Dakota'
})

# Add presentations and publications
for pub in research_profile['publications_presentations']:
    timeline_data.append({
        'date': f"{pub['year']}-06-01",  # Approximate dates
        'event': pub['title'][:50] + "..." if len(pub['title']) > 50 else pub['title'],
        'type': pub['type'],
        'importance': 8 if pub['type'] == 'Conference Talk' else 6,
        'description': pub['venue']
    })

# Add project milestones
for i, project in enumerate(research_profile['featured_projects']):
    start_date = pd.Timestamp('2024-01-01') - pd.DateOffset(months=project['duration_months'])
    timeline_data.append({
        'date': start_date.strftime('%Y-%m-%d'),
        'event': f"Started: {project['title']}",
        'type': 'Project Start',
        'importance': 7,
        'description': project['description'][:80] + "..."
    })
    
    timeline_data.append({
        'date': '2024-01-01',
        'event': f"Completed: {project['title']}",
        'type': 'Project Completion',
        'importance': 9,
        'description': f"Status: {project['status']}"
    })

timeline_df = pd.DataFrame(timeline_data)
timeline_df['date'] = pd.to_datetime(timeline_df['date'])
timeline_df = timeline_df.sort_values('date')

print("ğŸ“… Research Timeline Analysis")
print(timeline_df[['date', 'event', 'type']].to_string(index=False))
```

```python
# Interactive Research Timeline Visualization

# Create timeline chart
fig = go.Figure()

# Color mapping for event types
color_map = {
    'Education': '#00a8e1',      # Prime blue
    'Conference Talk': '#ff6b6b',  # Red
    'Poster': '#4ecdc4',         # Teal  
    'Project Start': '#95e1d3',   # Light green
    'Project Completion': '#45b7d1'  # Blue
}

for event_type in timeline_df['type'].unique():
    df_subset = timeline_df[timeline_df['type'] == event_type]
    
    fig.add_trace(go.Scatter(
        x=df_subset['date'],
        y=df_subset['importance'],
        mode='markers+text',
        marker=dict(
            size=df_subset['importance'] * 2,
            color=color_map.get(event_type, '#gray'),
            symbol='circle',
            line=dict(width=2, color='white')
        ),
        text=df_subset['event'],
        textposition="top center",
        name=event_type,
        hovertemplate='<b>%{text}</b><br>' +
                     'Date: %{x}<br>' +
                     'Type: ' + event_type + '<br>' +
                     '<extra></extra>'
    ))

fig.update_layout(
    title="Deepika Nuthalapati - Research & Academic Timeline",
    title_x=0.5,
    xaxis_title="Timeline",
    yaxis_title="Impact/Importance",
    height=500,
    hovermode='closest',
    paper_bgcolor='rgba(15, 23, 30, 1)',
    plot_bgcolor='rgba(15, 23, 30, 0.8)',
    font=dict(color='white', size=12),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    )
)

fig.update_xaxes(
    gridcolor='rgba(255, 255, 255, 0.2)',
    showgrid=True
)

fig.update_yaxes(
    gridcolor='rgba(255, 255, 255, 0.2)',
    showgrid=True,
    range=[0, 12]
)

fig.show()

# Summary statistics
print("\nğŸ“Š Research Activity Summary:")
print(f"ğŸ“ Total Publications/Presentations: {len(research_profile['publications_presentations'])}")
print(f"ğŸ”¬ Featured Research Projects: {len(research_profile['featured_projects'])}")
print(f"ğŸ¯ Primary Research Areas: {len(research_profile['research_areas'])}")
print(f"ğŸ† GitHub Contributions (last year): {research_profile['github_stats']['total_contributions']}")

# Research impact metrics
venue_types = timeline_df['type'].value_counts()
print(f"\nğŸ“ˆ Research Output by Type:")
for venue, count in venue_types.items():
    print(f"   {venue}: {count}")
```
